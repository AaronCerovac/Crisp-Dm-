In this project we are going to use Naive Bayes classifier for crating a model.

Naive Bayes is a simple technique for constructing classifiers that are working by assigning class labels to elements and with that make classification and drawn some finite set. 
We are using this type of classifier because it has proven to not only be simple but also fast, accurate, and reliable when we have to work with this type of data set. 

Here is one simple example, suppose we are building a classifier that says whether a text is about weather or not.

Now, when we have some of out training data we can train model, which will use naive bayer classifier to calculate probability using example from out training data on new test data. It is easy to make classification if we have the same sentence as in training data, but it is becoming harder to tag a sentence if it is not in the training data, such as "The day will be sunny".

## 4.2. Modelling assumptions 
Naive Bayes classifier is with an assumption of independence among elements. 
In simple terms, a Naive Bayes classifier assumes that the presence of a particular feature in a class is unrelated to the presence of any other feature.

For example, a fruit may be considered to be an orange if it is orange, round, and about 2 or more inches in diameter.
Even if these features depend on each other or upon the existence of the other features, all of these elements of fruit independently play a part in the probability that this fruit is an orange.

For this algorithm to work, we will need to make the target variable binary.
This algorithm can work very good with missing values, but in our case there is no problem with that, since we don't have any missing values. 

## 4.3. Build Model

We are going to change our target variable and convert it from string to binary, since as we said this algorithm is working with binary data. 

```python
news['TrueOrFake'] = news.TrueOrFake.map({'True': 0, 'Fake': 1})
```

Looking at the first six rows in changed data

```python
news.head(6)
```

Looking at the last six rows in changed data

```python
news.tail(6)
```

Convert all characters to lowercase so the algorithm can make it easier to read, and we are doing this on whole data set

```python
news['title'] = news.title.map(lambda x: x.lower())
```

```python
news['text'] = news.text.map(lambda x: x.lower())
```

Looking at the changed data in our data set

```python
news.head(6)
```

Using this line of code, we will remove punctuation in our text so that there is no difference between the words

```python
news['title'] = news.title.str.replace('[^\w\s]', '')
```

```python
news['text'] = news.text.str.replace('[^\w\s]', '')
```

Looking at the changed data in our data set

```python
news.head(6)
```

Importing library for tokenizing

```python
import nltk
```

```python
nltk.download('punkt')
```

Using this line of code, we will doing tokenization process.
Tokenization is the process of replacing sensitive data with unique identification, simply put and in our case we tokenize each word and change it to a unique number so that the algorithm can make it easier and faster to read true data.

```python
news['title'] = news['title'].apply(nltk.word_tokenize)
```

```python
news['text'] = news['text'].apply(nltk.word_tokenize)
```

Importing library for stemming.
Stemming is basically removing the suffix from a word and reduce it to its root word, for example “Flying” is a word and its suffix is “ing”, if we remove “ing” from “Flying” then we will get base word or root word which is “Fly”.
We do that so that the whole process of modeling with this algorithm can go faster.


```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
news['title'] = news['title'].apply(lambda x: [stemmer.stem(y) for y in x])
```

```python
news['text'] = news['text'].apply(lambda x: [stemmer.stem(y) for y in x])
```

```python
news.head(6)
```

Importing library and converts the list of words into space-separated strings.
Simple put, separated each word from the next by a single "," (comma) character.
Each word itself must not begin or end with any space characters, and each word itself must not contain any "," (comma) characters.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
news['title'] = news['title'].apply(lambda x: ' '.join(x))
```

```python
news['text'] = news['text'].apply(lambda x: ' '.join(x))
```

```python
count_vect = CountVectorizer()
```

```python
counts = count_vect.fit_transform(news['title'])
```

```python
counts = count_vect.fit_transform(news['text'])
```

Looking at the changed data in our data set

```python
news.head(6)
```

Importing library and making TFIDF, short for term frequency–inverse document frequency.
Term frequency is how many times a term appears in a particular document in your corpus. That information is needed for algorithm to work and to make calculation for final term decision.

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
transformer = TfidfTransformer().fit(counts)
```

```python
counts = transformer.transform(counts)
```

Importing library and making the process of splitting our data set into training and test set.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(counts, news['TrueOrFake'], test_size=0.1,
random_state=0)
```

Importing library and fitting Multinomial Naive Bayes Classifier on our data set. 

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB().fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
```

Looking at the accuracy of the model

```python
print("Accuracy of the model is:",np.mean(predicted == y_test),"%")
```

Creating the confusion matrix to see how did our model perform. 
A confusion matrix is a table that is often used to describe the performance of a classification model (or “classifier”) on a set of test data for which the true values are known.

```python
from sklearn.metrics import confusion_matrix
```

```python
cf_matrix = confusion_matrix(y_test, predicted)
print(cf_matrix)
```

Creating the confusion matrix that is little bet more visual but it is still showing all the data for our model and how did it perform.

```python
group_names = ["True Neg","False Pos","False Neg","True Pos"]
group_counts = ["{0:0.0f}".format(value) for value in
                cf_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cf_matrix.flatten()/np.sum(cf_matrix)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cf_matrix, annot=labels, fmt="", cmap='Blues')
```
